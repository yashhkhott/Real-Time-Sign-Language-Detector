```python
import cv2
import os
import time
import uuid
```

```python
labels = ['hello', 'thanks', 'yes', 'no', 'iloveyou']
number_imgs = 15
```

```python
base_path = os.path.join('Tensorflow', 'workspace', 'images', 'collectedimages')

for label in labels:
    label_path = os.path.join(base_path, label)
    os.makedirs(label_path, exist_ok=True)

    cap = cv2.VideoCapture(0)  

    print('Collecting images for {}'.format(label))
    time.sleep(5)

    for imgnum in range(number_imgs):
        ret, frame = cap.read()

        if not ret:
            print("Failed to capture image")
            continue
        
        imgname = os.path.join(label_path, f'{label}.{str(uuid.uuid1())}.jpg')
        success = cv2.imwrite(imgname, frame)
        
        if success:
            print(f"Image saved at {imgname}")
        else:
            print(f"Failed to save image at {imgname}")

        cv2.imshow('frame', frame)
        time.sleep(2)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

    
```
